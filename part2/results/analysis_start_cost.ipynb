```python
import pandas as pd

# Initialize lists to store the extracted data
times = []
minimum_costs = []

# Loop through each file index
for i in range(1, 28):  # This will loop from 1 to 26
    # Construct the file name based on the current index
    file_name = f'Duration{26}_start{i-1}.csv'
    # Read the CSV file
    df = pd.read_csv(file_name)
    
    # Extract the 'time' and 'minimum_cost' from the first row of the DataFrame
    time = df['time'][0]
    minimum_cost = df['minimum_cost'][0]
   
    
    # Append the extracted data to the lists
    times.append(time)
    minimum_costs.append(minimum_cost)

# Optionally, you can create a DataFrame to store all the results together
results_df = pd.DataFrame({
    'time': times,
    'minimum_cost': minimum_costs
})

# If you want to save this data to a new CSV file
results_df.to_csv('summary_results.csv', index=False)

# To display or use the collected data
print(results_df)
```

```python
# plot the results_df
plt.plot(results_df['time'], results_df['minimum_cost'])
plt.xlabel('start time')
plt.ylabel('minimum_cost')
plt.title('minimum_cost vs start time')

plt.show()

```
