```python
import os 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cur_path = os.getcwd()

path_3060 = os.path.join(cur_path, '3060_95W/Data')
path_3060_alexnet = os.path.join(path_3060, 'alexnet/fashion_mnist/E10_B128_R0')
path_3060_resnet = os.path.join(path_3060, 'resnet/fashion_mnist/E10_B128_R0')
print(path_3060_alexnet, '\n', path_3060_resnet) 

path_4090 = os.path.join(cur_path, '4090/Data')
path_4090_alexnet = os.path.join(path_4090, 'alexnet/fashion_mnist/E10_B128_R0')
path_4090_resnet = os.path.join(path_4090, 'resnet/fashion_mnist/E10_B128_R0')
print(path_4090_alexnet, '\n', path_4090_resnet)
```

```python
# show the files in the directory
folder_names = [path_3060_alexnet, path_3060_resnet, path_4090_alexnet, path_4090_resnet]
print(folder_names)
for name in folder_names:
    print(os.listdir(name))
```

```python
alexnet_3060_energy = np.load(os.path.join(path_3060_alexnet, 'Epoch_energy.npy'), allow_pickle=True)
resnet_3060_energy = np.load(os.path.join(path_3060_resnet, 'Epoch_energy.npy'), allow_pickle=True)
alexnet_4090_energy = np.load(os.path.join(path_4090_alexnet, 'Epoch_energy.npy'), allow_pickle=True)
resnet_4090_energy = np.load(os.path.join(path_4090_resnet, 'Epoch_energy.npy'), allow_pickle=True)
# print(alexnet_3060_energy)
```

```python
def epochinfo(energyfile):
    epoch_energy_info = []
    for i in range(len(energyfile)):
        epochi_energy = energyfile[i][0]
        epochi_totalenergy = np.sum(epochi_energy)
        epochi_totalenergy = np.round(epochi_totalenergy, 2)
        epoch_energy_info.append(epochi_totalenergy)
    return epoch_energy_info
```

```python
alexnet_3060_energy_info = epochinfo(alexnet_3060_energy)
resnet_3060_energy_info = epochinfo(resnet_3060_energy)
alexnet_4090_energy_info = epochinfo(alexnet_4090_energy)
resnet_4090_energy_info = epochinfo(resnet_4090_energy)
print(alexnet_3060_energy_info)
print(resnet_3060_energy_info)
print(alexnet_4090_energy_info)
print(resnet_4090_energy_info)
```

```python
# plot the energy consumption separately based on the model in 2 subplots， each subplot contains 2 lines with 2 GPUs in the same model
# plot the each value of energy consumption in each epoch
fig, axs = plt.subplots(2, 1, figsize=(10, 10))
fig.suptitle('Energy Consumption of AlexNet and ResNet on Fashion MNIST')
axs[0].plot(alexnet_3060_energy_info, label='3060_AlexNet')
axs[0].plot(alexnet_4090_energy_info, label='4090_AlexNet')
axs[0].set_title('AlexNet')
axs[0].set_xlabel('Epoch')
axs[0].set_ylabel('Energy Consumption (J)')
for i in range(len(alexnet_3060_energy_info)):
    axs[0].text(i, alexnet_3060_energy_info[i], alexnet_3060_energy_info[i])
axs[0].legend()

axs[1].plot(resnet_3060_energy_info, label='3060_ResNet')
axs[1].plot(resnet_4090_energy_info, label='4090_ResNet')
axs[1].set_title('ResNet')
axs[1].set_xlabel('Epoch')
axs[1].set_ylabel('Energy Consumption (J)')
for i in range(len(resnet_3060_energy_info)):
    axs[1].text(i, resnet_3060_energy_info[i], resnet_3060_energy_info[i])
axs[1].legend()

plt.show()
```

```python
# Plotting the comparison of energy consumption for each model on both hardware devices with more distinct colors
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 10), sharex=True)

# Comparison for AlexNet
ax1.plot(alexnet_3060_energy_info, label='3060 AlexNet', marker='o', color='blue')
ax1.plot(alexnet_4090_energy_info, label='4090 AlexNet', marker='o', color='red')
ax1.set_title('Energy Consumption: AlexNet')
ax1.set_ylabel('Energy Consumption')
ax1.legend()

# Comparison for ResNet
ax2.plot(resnet_3060_energy_info, label='3060 ResNet', marker='o', color='green')
ax2.plot(resnet_4090_energy_info, label='4090 ResNet', marker='o', color='purple')
ax2.set_title('Energy Consumption: ResNet')
ax2.set_xlabel('Index')
ax2.set_ylabel('Energy Consumption')
ax2.legend()

plt.tight_layout()
plt.show()
```

```python
# Plotting the comparison of energy consumption for each model on both hardware devices with data points labeled
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 10), sharex=True)

# Comparison for AlexNet
ax1.plot(alexnet_3060_energy_info, label='3060 AlexNet', marker='o', color='blue')
ax1.plot(alexnet_4090_energy_info, label='4090 AlexNet', marker='o', color='red')
for i, txt in enumerate(alexnet_3060_energy_info):
    ax1.annotate(txt, (i, alexnet_3060_energy_info[i]), textcoords="offset points", xytext=(0,5), ha='center')
for i, txt in enumerate(alexnet_4090_energy_info):
    ax1.annotate(txt, (i, alexnet_4090_energy_info[i]), textcoords="offset points", xytext=(0,5), ha='center')
ax1.set_title('Energy Consumption: AlexNet')
ax1.set_ylabel('Energy Consumption')
ax1.legend()

# Comparison for ResNet
ax2.plot(resnet_3060_energy_info, label='3060 ResNet', marker='o', color='green')
ax2.plot(resnet_4090_energy_info, label='4090 ResNet', marker='o', color='purple')
for i, txt in enumerate(resnet_3060_energy_info):
    ax2.annotate(txt, (i, resnet_3060_energy_info[i]), textcoords="offset points", xytext=(0,5), ha='center')
for i, txt in enumerate(resnet_4090_energy_info):
    ax2.annotate(txt, (i, resnet_4090_energy_info[i]), textcoords="offset points", xytext=(0,5), ha='center')
ax2.set_title('Energy Consumption: ResNet')
ax2.set_xlabel('Index')
ax2.set_ylabel('Energy Consumption')
ax2.legend()

plt.tight_layout()
plt.show()
```

```python
alexnet_3060_energy_avg = np.mean(alexnet_3060_energy_info)
alexnet_4090_energy_avg = np.mean(alexnet_4090_energy_info)
differ_alexnet = alexnet_4090_energy_avg - alexnet_3060_energy_avg
resnet_3060_energy_avg = np.mean(resnet_3060_energy_info)
resnet_4090_energy_avg = np.mean(resnet_4090_energy_info)
differ_resnet = resnet_4090_energy_avg - resnet_3060_energy_avg
print('Average Energy Consumption of AlexNet on 3060:', differ_alexnet)
print('Average Energy Consumption of AlexNet on 4090:', differ_resnet)
```
