```python
import os 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cur_path = os.getcwd()

path_4090 = os.path.join(cur_path, '4090/Data')
path_alexnet_f = os.path.join(path_4090, 'alexnet_AllLayers/fashion_mnist/E20_B256_R0')
path_resnet_f = os.path.join(path_4090, 'resnet18/fashion_mnist/E20_B256_R0')
path_vgg11_f = os.path.join(path_4090, 'vgg11/fashion_mnist/E20_B256_R0')
path_googlenet_f = os.path.join(path_4090, 'googlenet/fashion_mnist/E20_B256_R0')

path_alexnet_c = os.path.join(path_4090, 'alexnet_AllLayers/cifar100/E20_B256_R0')
path_resnet_c = os.path.join(path_4090, 'resnet18/cifar100/E20_B256_R0')
path_vgg11_c = os.path.join(path_4090, 'vgg11/cifar100/E20_B256_R0')
path_googlenet_c = os.path.join(path_4090, 'googlenet/cifar100/E20_B256_R0')

print(path_alexnet_f,'\n', path_resnet_f,'\n', path_vgg11_f,'\n', path_googlenet_f,'\n',
        path_alexnet_c,'\n', path_resnet_c,'\n', path_vgg11_c,'\n', path_googlenet_c)
```

```python
alexnet_f = np.load(os.path.join(path_alexnet_f, 'Epoch_energy.npy'), allow_pickle=True)
alexnet_c = np.load(os.path.join(path_alexnet_c, 'Epoch_energy.npy'), allow_pickle=True)

resnet_f = np.load(os.path.join(path_resnet_f, 'Epoch_energy.npy'), allow_pickle=True)
resnet_c = np.load(os.path.join(path_resnet_c, 'Epoch_energy.npy'), allow_pickle=True)

vgg11_f = np.load(os.path.join(path_vgg11_f, 'Epoch_energy.npy'), allow_pickle=True)
vgg11_c = np.load(os.path.join(path_vgg11_c, 'Epoch_energy.npy'), allow_pickle=True)

googlenet_f = np.load(os.path.join(path_googlenet_f, 'Epoch_energy.npy'), allow_pickle=True)
googlenet_c = np.load(os.path.join(path_googlenet_c, 'Epoch_energy.npy'), allow_pickle=True)
```

```python
def epochinfo(energyfile):
    epoch_energy_info = []
    for i in range(len(energyfile)):
        epochi_energy = energyfile[i][0]
        epochi_totalenergy = np.sum(epochi_energy)
        epochi_totalenergy = np.round(epochi_totalenergy, 2)
        epoch_energy_info.append(epochi_totalenergy)
    return epoch_energy_info
```

```python
alexnet_f_info = epochinfo(alexnet_f)
alexnet_c_info = epochinfo(alexnet_c)

resnet_f_info = epochinfo(resnet_f)
resnet_c_info = epochinfo(resnet_c)

vgg11_f_info = epochinfo(vgg11_f)
vgg11_c_info = epochinfo(vgg11_c)

googlenet_f_info = epochinfo(googlenet_f)
googlenet_c_info = epochinfo(googlenet_c)

print('AlexNet Fashion:', alexnet_f_info, '\nAlexNet CIFAR:', alexnet_c_info)
print('ResNet Fashion:', resnet_f_info, '\nResNet CIFAR:', resnet_c_info)
print('VGG11 Fashion:', vgg11_f_info, '\nVGG11 CIFAR:', vgg11_c_info)
print('GoogLeNet Fashion:', googlenet_f_info, '\nGoogLeNet CIFAR:', googlenet_c_info)
```

```python
# Plotting the comparison of energy consumption for each model on both hardware devices with data points labeled
fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1, figsize=(12, 10), sharex=True)

# Comparison for AlexNet
ax1.plot(alexnet_f_info, label='FashionMNIST', marker='o', color='blue')
ax1.plot(alexnet_c_info, label='Cifar100', marker='o', color='red')
for i, txt in enumerate(alexnet_f_info):
    ax1.annotate(txt, (i, alexnet_f_info[i]), textcoords="offset points", xytext=(0,5), ha='center')
for i, txt in enumerate(alexnet_c_info):
    ax1.annotate(txt, (i, alexnet_c_info[i]), textcoords="offset points", xytext=(0,5), ha='center')
ax1.set_title('Energy Consumption: AlexNet')
ax1.set_ylabel('Energy Consumption')
ax1.legend()

# Comparison for ResNet
ax2.plot(resnet_f_info, label='FashionMNIST', marker='o', color='green')
ax2.plot(resnet_c_info, label='Cifar100', marker='o', color='purple')
for i, txt in enumerate(resnet_f_info):
    ax2.annotate(txt, (i, resnet_f_info[i]), textcoords="offset points", xytext=(0,5), ha='center')
for i, txt in enumerate(resnet_f_info):
    ax2.annotate(txt, (i, resnet_c_info[i]), textcoords="offset points", xytext=(0,5), ha='center')
ax2.set_title('Energy Consumption: ResNet')
ax2.set_xlabel('Index')
ax2.set_ylabel('Energy Consumption')
ax2.legend()

ax3.plot(vgg11_f_info, label='FashionMNIST', marker='o', color='orange')
ax3.plot(vgg11_c_info, label='Cifar100', marker='o', color='brown')
for i, txt in enumerate(vgg11_f_info):
    ax3.annotate(txt, (i, vgg11_f_info[i]), textcoords="offset points", xytext=(0,5), ha='center')
for i, txt in enumerate(vgg11_c_info):
    ax3.annotate(txt, (i, vgg11_c_info[i]), textcoords="offset points", xytext=(0,5), ha='center')
ax3.set_title('Energy Consumption: VGG11')
ax3.set_xlabel('Index')
ax3.set_ylabel('Energy Consumption')
ax3.legend()

ax4.plot(googlenet_f_info, label='FashionMNIST', marker='o', color='pink')
ax4.plot(googlenet_c_info, label='Cifar100', marker='o', color='black')
for i, txt in enumerate(googlenet_f_info):
    ax4.annotate(txt, (i, googlenet_f_info[i]), textcoords="offset points", xytext=(0,5), ha='center')
for i, txt in enumerate(googlenet_c_info):
    ax4.annotate(txt, (i, googlenet_c_info[i]), textcoords="offset points", xytext=(0,5), ha='center')
ax4.set_title('Energy Consumption: GoogLeNet')
ax4.set_xlabel('Index')
ax4.set_ylabel('Energy Consumption')
ax4.legend()

plt.tight_layout()
plt.show()
```

```python
alexnet_f_avg = np.mean(alexnet_f_info)
alexnet_c_avg = np.mean(alexnet_c_info)

resnet_f_avg = np.mean(resnet_f_info)
resnet_c_avg = np.mean(resnet_c_info)

vgg11_f_avg = np.mean(vgg11_f_info)
vgg11_c_avg = np.mean(vgg11_c_info)

googlenet_f_avg = np.mean(googlenet_f_info)
googlenet_c_avg = np.mean(googlenet_c_info)

differ_alexnet = alexnet_f_avg - alexnet_c_avg
differ_resnet = resnet_f_avg - resnet_c_avg
differ_vgg11 = vgg11_f_avg - vgg11_c_avg
differ_googlenet = googlenet_f_avg - googlenet_c_avg

print('The difference in energy consumption between FashionMNIST and CIFAR100 for AlexNet:', differ_alexnet)
print('The difference in energy consumption between FashionMNIST and CIFAR100 for ResNet:', differ_resnet)
print('The difference in energy consumption between FashionMNIST and CIFAR100 for VGG11:', differ_vgg11)
print('The difference in energy consumption between FashionMNIST and CIFAR100 for GoogLeNet:', differ_googlenet)
```
